## Import Packages

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

## Adjust for Missing Data

```python
missing_values = ['---']
df=pd.read_csv('heatmapmatrix.csv',header=0,na_values = missing_values)
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

## Format Genes

```python
genes=[]
for i in df["genes"]:
    if i!='-':
        #print(i.split("//"))
        genes.append(i.split("//")[1])
len(genes)
```

```python
df["genes"]=genes
df=df.set_index(df.columns[0])
```

```python
df.head()
```

## Gather information for HeatMap Labels

```python
cols=[i for i in df.columns]
cols
```

```python
annotations=pd.read_csv('annotations.csv',header=0)
```

```python
annotations
```

## Heatmap Visualization. 

##### Contrast/Group by Non-AD and AD samples. 

##### All samples from the frontal cortex region

```python
colors=[]
for i in list(annotations["x"]):
    if i.find("non-AD")!=-1:
        colors.append("pink")
    else:
        colors.append("purple")
         
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,20))
ax.set_yticks(np.arange(len(genes))+0.5)
ax.set_yticklabels(genes)
ax.set_xticks(np.arange(len(cols))+0.5)
ax.set_xticklabels(cols, rotation=45)
plt.pcolormesh(df, edgecolors='k',cmap="bone",linewidth=0)
for ytick, color in zip(ax.get_yticklabels(), colors):
    ytick.set_color(color)

ax = plt.gca()
ax.set_aspect('equal')

import matplotlib.patches as mpatches
purple_patch = mpatches.Patch(color='purple', label='Alzheimers')
pink_patch = mpatches.Patch(color='pink', label='Non-Alzheimers')
plt.legend(handles=[purple_patch, pink_patch])
plt.savefig('frontalcortex.pdf')
```
