## Get Dataset from GEO Database:

```r
eset <- getGEO('GSE36980')[[1]]
#class is an ExpressionSet from the Biobase package
class(eset)
#33297 genes and 80 samples
show(eset)
# Dimensions of the ExpressionSet
#dim(eset)
#nrow(eset)
#ncol(eset)
#sampleNames(eset)[1:8]
#featureNames(eset)[1:8]


#fData is for genes! fData is a dataframe.
#class(fData(eset))
#dim(fData(eset))
##rows of FData should be the same as featureNames of expressionSet
all.equal(featureNames(eset),rownames(fData(eset)))
head(fData(eset))
#colnames(fData(eset))

#pData is also a dataframe.
#class(pData(eset))
#dim(pData(eset))
all.equal(sampleNames(eset),rownames(pData(eset)))
head(pData(eset))
#colnames(pData(eset))

```

```r
#Expression Data
head(exprs(eset))
#class(exprs(eset))
summary(exprs(eset))
all.equal(colnames(exprs(eset)),sampleNames(eset))
all.equal(rownames(exprs(eset)),featureNames(eset))

#can subset if needed
#eset[1:10,]
```

## Starting Data File Preparation for HeatMap Visualization

```r
table(pData(eset)$source_name_ch1)
```

```r
#Only get data for the given 2 brain regions
k <- which(pData(eset)$source_name_ch1=="Frontal cortex of AD brain" | pData(eset)$source_name_ch1=="Frontal cortex of non-AD brain")
type=pData(eset)[k,]


#summary(log2(exprs(eset)))
#hist(log2(exprs(eset)))
# Make the heatmap
#heatmap.2(exprs(AD),trace='none')

```

```r
type=pData(eset)[k,]
type
```

```r
#Expression Data filtered
exprs=exprs(eset)[,match(row.names(type),colnames(exprs(eset)))]
hist(exprs)
```

```r
##Top 50 genes. In this case, less than 50 genes available for this given brain region.
```

```r
#summary(exprs(AD))
stdDev <- apply(exprs,1,sd)
#hist(stdDev)
exprs = exprs[order(stdDev,decreasing=TRUE)[1:50],]
```

```r
genes=fData(eset)[match(rownames(exprs), rownames(fData(eset))),"gene_assignment"]
```
